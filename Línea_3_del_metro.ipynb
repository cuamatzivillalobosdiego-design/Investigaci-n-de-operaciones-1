Rehabilitación para la linea 3 del metro de la CDMX.



Los costos de quiebra son uno de los factores más importantes que afectan las decisiones de financiación corporativas. Son un parámetro crucial tanto en estudios relacionados con la estructura de capital como en los modelos de fijación de precios de activos financieros en condiciones de default.

En este proyecto buscamos reducir el tiempo de finalizacion.

````python
import pandas as pd

# Crear un diccionario con datos
data = {
    "Actividad": ["A", "B", "C", "D", "E", "F", "G", "H"],
    "Descripcion de la actividad": ["Ruta de RTP", "Renivelacion y reforzamiento",
                                    "Modernizacion de sistemas", "Instalacion de seguridad",
                                    "Mantenimiento mayor y rehabilitacion", "Adquisicion de nuevos trenes",
                                    "Renovacion estetica y funcional", "Mejora de servicio"],
    "Predecesores inmediatos": ["---", "A", "B", "B", "D", "C,E", "C,F", "G"],
    "Duracion estimada (Semanas)": ["2", "14", "12", "7", "8", "3", "14", "10"]
}

# Convertir en DataFrame (tabla)
df = pd.DataFrame(data)

# Mostrar la tabla
df
````

````python

import networkx as nx
import matplotlib.pyplot as plt

Q = nx.DiGraph() #grafo dirigido

Q.add_weighted_edges_from(
    [
        ("A", "B", 2),
        ("B", "C", 14),
        ("B", "D", 14),
        ("D", "E", 7),
        ("C", "F", 12),
        ("C", "G", 12),
        ("E", "F", 8),
        ("F", "G", 3),
        ("G", "H", 14)
    ])
# Definir layout (kamada_kawai)
pos = nx.kamada_kawai_layout(Q)

fig, ax = plt.subplots(figsize=(8, 6))  # más grande para ver bien
nx.draw(Q, pos, with_labels=True, node_size=200, node_color="skyblue", arrows=True, ax=ax)

# Escalar límites del gráfico
ax.set_xlim([1.2*x for x in ax.get_xlim()])
ax.set_ylim([1.2*y for y in ax.get_ylim()])

plt.title("Red de proyecto de rehabilitacion para la linea 3 del metro ")
plt.show()
````

````python
nx.shortest_path(Q, source="A", target="H")
````
