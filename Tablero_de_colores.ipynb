```python
import matplotlib.pyplot as plt
import numpy as np

x = 1
y = 0
z = 0
t = 0
n = 15
```

```python
#Matriz de transicion
matriz = np.array([[0.25, 0, 0.25, 0.5],
                   [0.5, 0.25, 0, 0.25],
                   [0.25, 0.5, 0.25, 0],
                   [0, 0.25, 0.5, 0.25]])
matriz_n = np.linalg.matrix_power(matriz, n)
matriz_n
```

```python
#Matriz de valores iniciales
matriz_x = np.array([[x],
                     [y],
                     [z],
                     [t]])
#Multiplicacion matriz elevada a n con matriz de valores iniciales
matriz_resultado= np.dot(matriz_n, matriz_x)
#Convergencia [x, y, z, t] =
matriz_resultado
```

```python
estados = [matriz_x.flatten()]

for i in range(1, n + 1):
    matriz_x = np.dot(matriz, matriz_x)
    estados.append(matriz_x.flatten())

# Convertimos a array para fácil manejo
estados = np.array(estados)

# Grafica
plt.figure(figsize=(8,5))
plt.plot(estados[:, 0], label="Azul", marker='o', markersize=3)
plt.plot(estados[:, 1], label="Naranja", marker='o', markersize=3)
plt.plot(estados[:, 2], label="Verde", marker='o', markersize=3)
plt.plot(estados[:, 3], label="Rojo", marker='o', markersize=3)
plt.xlabel("Número de pasos (n)")
plt.ylabel("Probabilidad")
plt.title("Convergencia al estado estacionario")
plt.legend()
plt.grid(True)
plt.show()
```
